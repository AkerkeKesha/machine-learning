4. Постройте heatmap для матрицы корреляций, есть ли признаки, которые сильно скоррелированы? Какие это признаки?


```python
Corr = df.corr()
```

```python
plt.figure(figsize=(15,8));
sns.heatmap(Corr, annot=True, fmt='.1f');
```

```python
CorField = []
for i in Corr:
    for j in Corr.index[Corr[i] > 0.8]:
        if i != j and j not in CorField and i not in CorField:
            CorField.append(j)
            print (i, j, Corr[i][Corr.index == j].values[0])
```

Сильно между собой скоррелированы (больше 0.8):
1. radius_mean с perimeter_mean, area_mean, concave points_mean, radius_worst, perimeter_worst, area_worst
2. Еще texture_mean c texture_worst
3. Также smoothness_mean c smoothness_worst
4. compactness_mean c признаками concavity_mean, compactness_worst, concavity_worst, concave points_worst
5. radius_se perimeter_se, area_se
6. compactness_se concavity_se, compactness_se, fractal_dimension_se

5. Постройте для сильно скоррелированных признаков попарные scatterplot-ы, действительно ли наблюдается линейная зависимость?


```python
sns.pairplot(df[['radius_mean', 'perimeter_mean']])
#есть очевидная линейная зависимость
```

```python
sns.pairplot(df[['radius_mean', 'area_mean']])
# area is radius squared. so there is paraboloid correlation not linear
# неочевидная линейная зависимость. видна парабола. 
```

```python
sns.pairplot(df[['radius_mean', 'concave points_mean']])
# линейная зависимость видна
```

```python
sns.pairplot(df[['texture_mean', 'texture_worst']])
#texture mean включает в себя texture_worst поэтому такая корреляция
# линейная зависимость
```

```python
sns.pairplot(df[['compactness_mean', 'concavity_mean']])
#очевидная линейная зависимость
```

```python
sns.pairplot(df[['compactness_mean', 'concavity_mean']])
#очевидная линейная зависимость
```

6. Используя boxplots и группировку по целевой переменной, попробуйте предположить, по каким признакам наиболее удобно было бы отделить злокачественные новообразования от доброкачественных

```python
sns.boxplot(x='radius_mean', y='diagnosis', data = df)
```

У злокачественных опухолей средний радиус больше

```python
sns.boxplot(x='texture_mean', y='diagnosis', data = df)
```

```python
У злокачественных опухолей texture_mean(отклонение серого) тоже больше
```

```python
sns.boxplot(x='smoothness_mean', y='diagnosis', data = df)
```

```python
У злокачественных опухолей smoothness также бывает в среднем больше
```

Предпологаю, что более ли менее "независимые" признаки, а их около 4 из 33: radius_mean, texture_mean, smoothness_mean, compactness_mean. При построении модели удобно было бы предроложить используя эти признаки.
